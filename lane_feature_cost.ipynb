## Cost
#### 面积损失

```python
x = np.arange(6000)
y = np.array([])
for xx in x:
    if xx > 500:        
        yy = np.power(float(xx - 500)/250, 2) * 10
    else:
        yy = np.exp(np.abs(float(500-xx)/100)) *8
    y = np.append(y, yy)

plt.figure()
plt.plot(x[:3500],y[:3500])
plt.show()
```

#### 长度损失

```python
x = np.arange(1000)
y = np.array([])
for xx in x:
    if xx > 50:        
        yy = np.power(float(xx - 50)/80, 2)*20
    else:
        yy = np.exp(np.abs(float(50-xx)/10)) * 10
    y = np.append(y, yy)

plt.figure()
plt.plot(x[1:700],y[1:700])
plt.show()
```

### 宽度损失

```python
x = np.arange(20)
y = np.array([])
for xx in x:    
    yy = np.exp(np.abs(xx-12))/2
    y = np.append(y, yy)

plt.figure()
plt.plot(x[5:],y[5:])
plt.show()
```

#### 方向损失

```python
x = np.linspace(0, math.pi, 1000)
y = np.array([])
for xx in x:
#     xx = xx - math.pi /2  
#     if(xx > math.pi):
#         xx -= math.pi
#     if(xx < -math.pi):
#         xx = 2*math.pi + xx
    yy = np.power(np.abs(np.abs(xx - math.pi/2) - 0.7), 2)*1000
    y = np.append(y, yy)

plt.figure()
plt.plot(x,y)
plt.show()
```

#### 位置损失

```python
x = np.arange(720)
y = np.array([])
for xx in x:    
    yy = np.power(float(xx)/720, 3) * 1000
    y = np.append(y, yy)

plt.figure()
plt.plot(x,y)
plt.show()

x = np.arange(1204)
y = np.array([])
for xx in x:    
    yy = np.power(float(xx-620)/620, 2) * 1000
    y = np.append(y, yy)
plt.figure()
plt.plot(x,y)
plt.show()
```

#### 车道线位置损失

```python
x = np.arange(1024)
y = np.array([])
for xx in x:    
    yy = np.abs(np.mod(np.abs(xx-512), 192) - 96) * 4
    y = np.append(y, yy)
plt.figure()
plt.plot(x[224:800],y[224:800])
plt.show()
```

#### 填充度

```python
x = np.linspace(0.1, 1, 99)
y = np.array([])
for xx in x:    
    yy = -np.log2(xx) * 300
    y = np.append(y, yy)

plt.figure()
plt.plot(x,y)
plt.show()
```
